### 1. 以下の2つの伝達関数についてステップ応答を描画し，その違いを考察せよ (グラフは重ねなくてよい)．$t$の範囲は0sから5sまでで100点取ることとする．

- $P_1(s)$：零点なし $z = []$, 極 $p = -1\pm5j$, ゲイン $k = 1$
- $P_2(s)$：零点なし $z = []$, 極 $p = +1\pm5j$, ゲイン $k = 1$

### 2. 以下の3つの伝達関数についてステップ応答を描画し，その違いを考察せよ (グラフは重ねる) ．$t$の範囲は0sから5sまでで100点取ることとする．

- $P_3(s)$：零点なし $z = []$, 極 $p = -1\pm5j$, ゲイン $k = 26$
- $P_4(s)$：零点なし $z = []$, 極 $p = -3\pm5j$, ゲイン $k = 34$
- $P_5(s)$：零点なし $z = []$, 極 $p = -5\pm5j$, ゲイン $k = 50$

```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
```

```python
t = np.linspace(0, 5, 100)
```

###  $P_1(s)$：零点なし $z = []$, 極 $p = -1\pm5j$, ゲイン $k = 1$

```python
z1 = []
p1 = [-1+5j, -1-5j]
k1 = 1
n1, d1 = zpk2tf(z1, p1, k1)
P1 = tf(n1, d1)
```

```python
y1, t = step(P1, t)
```

```python
plt.plot(t, y1, color="blue")
plt.xlabel("t [s]")
plt.ylabel("y")
plt.grid()
plt.show()
```

### $P_2(s)$：零点なし $z = []$, 極 $p = +1\pm5j$, ゲイン $k = 1$

```python
z2 = []
p2 = [1+5j, 1-5j]
k2 = 1
n2, d2 = zpk2tf(z2, p2, k2)
P2 = tf(n2, d2)
```

```python
y2, t = step(P2, t)
```

```python
plt.plot(t, y2, color="blue")
plt.xlabel("t [s]")
plt.ylabel("y")
plt.grid()
plt.show()
```

### 1. 考察

極の実数部が負の場合は振動が次第に緩やかになり，安定する．

極の実数部が正の場合は振動が激しくなり，発散する．

上記のことから極の実数部の正負がシステムの安定性に関係していると考えられる．

### ・$P_3(s)$：零点なし $z = []$, 極 $p = -1\pm5j$, ゲイン $k = 26$
### ・$P_4(s)$：零点なし $z = []$, 極 $p = -3\pm5j$, ゲイン $k = 34$
### ・$P_5(s)$：零点なし $z = []$, 極 $p = -5\pm5j$, ゲイン $k = 50$

```python
z = []

p3 = [-1+5j, -1-5j]
k3 = 26
n3, d3 = zpk2tf(z, p3, k3)
P3 = tf(n3, d3)

p4 = [-3+5j, -3-5j]
k4 = 34
n4, d4 = zpk2tf(z, p4, k4)
P4 = tf(n4, d4)

p5 = [-5+5j, -5-5j]
k5 = 50
n5, d5 = zpk2tf(z, p5, k5)
P5 = tf(n5, d5)
```

```python
y3, t = step(P3, t)
y4, t = step(P4, t)
y5, t = step(P5, t)
```

```python
plt.plot(t, y3, label="$p = -1\pm5j$, $k = 26$", color="blue")
plt.plot(t, y4, label="$p = -3\pm5j$, $k = 34$", color="green", linestyle="--")
plt.plot(t, y5, label="$p = -5\pm5j$, $k = 50$", color="red", linestyle="-.")
plt.xlabel("t [s]")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

### 2. 考察

グラフより，極の実数部が小さいほど振動の収束が早くなっているため，極の実数部が負の場合，小さくなるほど収束が早くなると考えられる．

1の考察を含め，極の値がシステムの安定性に大きく関係していることが分かる．
