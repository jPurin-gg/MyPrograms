```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
```

### (1). 1次システム$P_{1st}(s) = 1/(1 +Ts)$の$\omega T= 1$のときのゲインの値[dB]および$\omega T= 0.2, 5$のときの位相の値[deg.]を計算せよ．割り切れない場合は小数点以下第1位まででよい．

・$\omega T=1$のゲイン[dB]
\begin{eqnarray}
20 \log_{10}|P_{1st}(s)| &=& 20\log_{10} \frac{1}{|1 + j\omega T|} \nonumber \\ \nonumber
&=& 20\log_{10} \frac{1}{\sqrt{1 + (\omega T)^2}}\\ \nonumber
&=& -10\log_{10} 2\\ \nonumber
&\approx& -3.01 
\end{eqnarray}

・$\omega T=0.2$の位相[deg]
\begin{eqnarray}
\angle P_{1st}(s) &=& -\tan^{-1}(\omega T) \nonumber \\ \nonumber
&=& -\tan^{-1}(0.2)\\ \nonumber
&\approx& -11.3 
\end{eqnarray}

・$\omega T=5$の位相[deg]
\begin{eqnarray}
\angle P_{1st}(s) &=& -\tan^{-1}(\omega T) \nonumber \\ \nonumber
&=& -\tan^{-1}(5)\\ \nonumber
&\approx& -26.6 
\end{eqnarray}

### (2). 2.  以下の信号$u(t)$を1次システム$P_1(s) = 1/(0.1s+1)$に入力したときの応答$y_1(t)$のグラフを描き，$u(t)$のグラフと比較してローパス特性を確認せよ．$t$の範囲は0sから4sとする．

\begin{equation}
u(t) = \sin(5t)+\sin(50t) \nonumber
\end{equation}

```python
t = np.linspace(0, 4, 100)

n = [1]
d = [0.1, 1]
P = tf(n, d)

u = np.sin(5*t)+np.sin(50*t)
```

```python
y, t, x = lsim(P, u, t)
```

```python
plt.plot(t, u, linestyle="--", label="u")
plt.plot(t, y, linestyle="-", label="y")
plt.legend()
plt.xlabel("t [s]")
plt.show()
```

上のグラフより，周波数の低い成分はそのまま出力されているが，周波数の高い成分は入力よりも小さくなっていることがわかる．

よってローパス特性を確認することができる．

### (3). 1次システム$P_1(s)=1/(0.1s+1)$のボード線図を書き，ゲイン線図と位相線図について，折れ線近似と重ねたグラフを作成せよ．

```python
n = [1]
d = [0.1, 1]
P = tf(n, d)
w = np.logspace(-3, 3, 100)

gain, phase, w = bode(P, w, plot=False)
g_dB = 20 * np.log10(gain)
p_deg = phase * 180 / np.pi

#ωT = 1で折れる
x1 = [0, 10, 1000]
y1 = [0, 0, -40]

#ωT = 0.2, 5で折れる
x2 = [0, 2, 50, 1000]
y2 = [0, 0, -90, -90]
```

```python
plt.semilogx(w, g_dB, label="bode", color="blue")
plt.plot(x1, y1, label="approximation", color="green", linestyle="--")
plt.xlabel("$\omega $ [rad/s]")
plt.ylabel("gain [dB]")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.semilogx(w, p_deg, label="bode", color="blue")
plt.plot(x2, y2, label="approximation", color="green", linestyle="--")
plt.xlabel("$\omega $ [rad/s]")
plt.ylabel("phase [deg.]")
plt.legend()
plt.grid()
plt.show()
```
