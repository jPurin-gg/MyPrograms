# 課題

一巡伝達関数が
\begin{eqnarray}
L(s)  = \frac{k}{s-1} \nonumber
\end{eqnarray}
次式で表されるシステムについて，Python-contorolを使って$K=2,0.75$の2通りのナイキスト線図を描き，ナイキストの安定判別法によりそれぞれの安定性を判別せよ．


また，閉ループ伝達関数$G_1(s)$の式を求め，安定性の必要十分条件（必要条件ではない）から上記の結果を確認せよ．

$L_1(s)$の極の中で実数部が正であるものの個数$U$は$U = 1$である．

```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
```

```python
z = []
p = [1]

k1 = 2
k2 = 0.75
```

### K = 2のナイキスト線図

```python
n, d = zpk2tf(z, p, k1)
L = tf(n, d)
real, imag, w = nyquist(L)
```

上記の結果より$N = 1$となる．

$U = 1$となることから，$U = N$が成り立つため，$K = 2$のとき$L_1(s)$は安定である．

### K = 0.75のナイキスト線図

```python
n, d = zpk2tf(z, p, k2)
L = tf(n, d)
real, imag, w = nyquist(L)
```

上記の結果より$N = 0$となる．

$U = 1$となることから，$U \neq N$となるため，$K = 0.75$のとき$L_1(s)$は不安定である．

閉ループ伝達関数$G_1(s)$を以下に示す．
\begin{eqnarray}
G_1(s)  &=& \frac{Y(s)}{U(s)} \nonumber \\
&=& \frac{L_1(s)}{1 + L_1(s)} \nonumber \\
&=& \frac{K}{(s-1)+K} \nonumber \\
&=& \frac{K}{s-(1-K)} \nonumber
\end{eqnarray}

安定性の必要十分条件から，$1-K < 0$となれば$L_1(s)$は安定する．

$K = 2$のとき
\begin{eqnarray}
1 - K = 1 - 2 = -1 < 0 \nonumber
\end{eqnarray}

$K = 0.75$のとき
\begin{eqnarray}
1 - K = 1 - 0.75 = 0.25 > 0 \nonumber
\end{eqnarray}

上記の結果から$K = 2$では安定となり，$K = 0.75$では不安定となる．

これはナイキストの安定判別法の結果と同じである．
