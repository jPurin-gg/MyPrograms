2.  以下の信号$u(t)$を1次システム$P_1(s) = 1/(0.1s+1)$に入力したときの応答$y_1(t)$のグラフを描き，$u(t)$のグラフと比較してローパス特性を確認せよ．$t$の範囲は0sから4sとする．

\begin{equation}
u(t) = \sin(5t)+\sin(50t) \nonumber
\end{equation}

3. 1次システム$P_1(s)=1/(0.1s+1)$のボード線図を書き，ゲイン線図と位相線図について，折れ線近似と重ねたグラフを作成せよ．

```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
```

### (1). 1次システム$P_{1st}(s) = 1/(1 +Ts)$の$\omega T= 1$のときのゲインの値[dB]および$\omega T= 0.2, 5$のときの位相の値[deg.]を計算せよ．割り切れない場合は小数点以下第1位まででよい．

・$\omega T=1$のゲイン[dB]
\begin{eqnarray}
20 \log_{10}|P_{1st}(s)| &=& 20\log_{10} \frac{1}{|1 + j\omega T|} \nonumber \\ \nonumber
&=& 20\log_{10} \frac{1}{\sqrt{1 + (\omega T)^2}}\\ \nonumber
&=& -10\log_{10} 2\\ \nonumber
&\approx& -3.01 
\end{eqnarray}

・$\omega T=0.2$の位相[deg]
\begin{eqnarray}
\angle P_{1st}(s) &=& -\tan^{-1}(\omega T) \nonumber \\ \nonumber
&=& -\tan^{-1}(0.2)\\ \nonumber
&\approx& -11.3 
\end{eqnarray}

・$\omega T=5$の位相[deg]
\begin{eqnarray}
\angle P_{1st}(s) &=& -\tan^{-1}(\omega T) \nonumber \\ \nonumber
&=& -\tan^{-1}(5)\\ \nonumber
&\approx& -26.6 
\end{eqnarray}

### (2). 2.  以下の信号$u(t)$を1次システム$P_1(s) = 1/(0.1s+1)$に入力したときの応答$y_1(t)$のグラフを描き，$u(t)$のグラフと比較してローパス特性を確認せよ．$t$の範囲は0sから4sとする．

\begin{equation}
u(t) = \sin(5t)+\sin(50t)
\end{equation}

```python
t = np.linspace(0, 4, 100)

n = [1]
d = [0.1, 1]
P = tf(n, d)

u = np.sin(5*t)+np.sin(50*t)
```

```python
y, t, x = lsim(P, u, t)
```

```python
plt.plot(t, u, linestyle="--", label="u")
plt.plot(t, y, linestyle="-", label="y")
plt.legend()
plt.xlabel("t [s]")
plt.show()
```

上のグラフよりローパス特性を確認することができる．

### (3). 1次システム$P_1(s)=1/(0.1s+1)$のボード線図を書き，ゲイン線図と位相線図について，折れ線近似と重ねたグラフを作成せよ．

```python
n = [1]
d = [0.1, 1]
P = tf(n, d)
w = np.logspace(-3, 3, 100)

gain, phase, w = bode(P, w, Plot=False)
g_dB = 20 * np.log10(gain)
p_deg = phase * 180 / np.pi

x1 = [0, 10, 1000]
x2 = [0, 2, 50, 1000]
```

```python
plt.semilogx(w, g_dB, label="ボード線図", color="blue")
plt.plot(x1, g_dB, label="折れ線近似", color="green", linestyle="--")
plt.xlabel("$\omega $ [rad/s]")
plt.ylabel("gain [dB]")
plt.grid()
plt.show()
```

```python
plt.semilogx(w, p_deg, label="ボード線図", color="blue")
plt.plot(x2, p_deg, label="折れ線近似", color="green", linestyle="--")
plt.xlabel("$\omega $ [rad/s]")
plt.ylabel("phase [deg.]")
plt.grid()
plt.show()
```
