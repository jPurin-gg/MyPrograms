```python
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
t = np.linspace(0, 30, 1000)
```

### (1) 演習4でプロットしたステップ応答について，定常偏差が0になっていることを確認し，さらに改善すべきはどんな点があるか考察せよ．プロットする範囲に気を付けること．

```python
Rm, Lm, M, Kx, Ki = 1, 1/30, 0.3, 1500, 3
KP, KI, KD = 1000, 100, 1000

n = [Ki*KD, Ki*KP, Ki*KI]
d = [M*Lm, M*Rm, Ki*KD-Kx*Lm, Ki*KP-Kx*Rm, Ki*KI]
P = tf(n, d)
y, t = step(P, t)
```

```python
plt.plot(t, y)
plt.xlabel("$t$ [s]")
plt.ylabel("Step response")
plt.grid()
plt.show()
```

上記のグラフより，定常偏差が0となっていることが確認できる．

今回の応答では，オーバーシュートが大きいため，これを小さくする必要がある．加えて，整定時間についても短くすることが理想的である．

以上が改善すべき点であると考えられる．

### (2) 演習4でステップ応答をプロットした磁気浮上系$G(s)$のランプ応答（$u(t) = t$としたときの応答）をプロットせよ．lsim関数を適切に使えばよい．入力と出力を重ねてプロットすること．

```python
u = t
y, t, x = lsim(P, u, t)
```

```python
plt.plot(t, u, linestyle="--", label="u")
plt.plot(t, y, linestyle="-", label="y")
plt.legend()
plt.xlabel("t [s]")
plt.ylabel("response")
plt.grid()
plt.show()
```

上記より，定常速度偏差が一定であることがわかる．

```python
print(y[999]-u[999])
```

また，定常速度偏差は約5.0となることがわかる．
